```python
import matplotlib.pyplot as plt

import random
nbPoints=int(input("Entrez le nombre de points : "))
dansCercle=0
x_axis_1,x_axis_2,y_axis_1,y_axis_2=[],[],[],[]
for i in range(nbPoints):
    x=random.random()
    y=random.random()
    if (x**2+y**2)**0.5<1:
        dansCercle+=1
        x_axis_1.append(x)
        y_axis_1.append(y)
    else:
        x_axis_2.append(x)
        y_axis_2.append(y)
print("Approximation de Ï€ :", dansCercle/nbPoints*4)

fig = plt.figure(1, figsize=(10, 10))
plt.scatter(x_axis_1,y_axis_1,s=20,c="blue")
plt.scatter(x_axis_2,y_axis_2,s=20,c="red")
plt.title("MÃ©thode de Monte-Carlo")
plt.show()
```
